Write a function find_missing that has three parameters, df, col1 and col2 where df is a DataFrame and
col1 and col2 are column names. This function should return all the rows of the DataFrame where col1
and col2 are missing. Only return the two columns as well. Answer problem 7 with this function.

```python
def find_missing(df, col1, col2):
    filt = df[col1].isna() & df[col2].isna()
    col = [col1, col2]
    return df.loc[filt, col]
```

```python
find_missing(movie, 'gross', 'budget')
```
